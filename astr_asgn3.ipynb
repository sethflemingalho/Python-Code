```python
x_plt = array([1])
y_plt = array([0])
vx_plt = array([0])
vy_plt = array([1])
t_plt = array([0])

## Execute Euler method
# set starting points for iteration to initial conditions
dt= .1*2*pi
tf=10*2*pi
x_old = 1
y_old = 0
vx = 0
vy = 1
t = 0# set starting time to 0


while (t <= tf):
    x = x_old + vx*dt        # implement Euler step
    y = y_old + vy*dt
    r = sqrt(x**2 + y**2)
    vx = vx - dt*(1*1*x)/r**3
    vy = vy - dt*(1*1*y)/r**3

    x_plt = append(x_plt,x)   # Append new points to the arrays
    y_plt = append(y_plt,y)
    vx_plt = append(vx_plt,vx)
    vy_plt = append(vy_plt,vy)
    t_plt = append(t_plt,t)

    x_old = x    # Update x_old and y_old for next Euler step
    y_old = y
    t = t + dt  # Increment time
```

```python
plt.figure(figsize=(6,6))
plot(x_plt,y_plt)
plt.xlabel("x")
plt.ylabel("y")
title("euler method")
```

The orbit is circular, but with the chosen timestep the orbit looks irregular/broken.

```python
plt.figure(figsize=(6,3))
Lz = x_plt*vy_plt - y_plt*vx_plt
plt.xlabel("time")
plt.ylabel("Lz")
title("Angular momentum vs time")
plot(t_plt,Lz)
ylim(-2,2)
```

Since Lz is a constant it is conserved as seen from the figure above
