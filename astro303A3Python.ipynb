```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#initial conditions

A =0.1

vx = 0
vy = 1
tinitial = 0
dt = A*np.pi*2 #controls the eccentricity
tf = 10*np.pi*2 #number of loops
xi = 1
yi = 0

xp = np.array([1])  
yp = np.array([0])
vxp = np.array([0])
vyp = np.array([1])
tp = np.array([0])


while (tinitial <= tf):
    x = xi + vx*dt
    y = yi +vy*dt
    R = np.sqrt(x**2 +y**2)
    vx = vx - dt*(x/R**3)
    vy = vy - dt*(y/R**3)
    
    xp = np.append(xp,x)
    yp = np.append(yp,y)
    vxp = np.append(vxp,vx)
    vyp = np.append(vyp,vy)
    tp = np.append(tp,tinitial)
    
    xi = x
    yi = y
    tinitial = tinitial+dt

# print(tp)
# print(xp)
# print(yp)
# print(vxp)
# print(vyp)

    
plt.plot(xp,yp)
plt.title('Euler Method')
```

```python
n = [1,2,3,4,5,6]
diffe1 = []

for i in range(len(n)):
    if i == 0:
        an = n[i]
        diffe1.append(0)
    else:
        an = n[i]


Energy = 0.5*((vxp)**2+(vyp)**2)-1/np.sqrt(xp**2 + yp**2)
diffe1 = []

for i in range(len(Energy)):
    if i ==0:
        an = Energy[i]
        diffe1.append(0)
    else:
        diffe1.append((Energy[i] - an)/an)
        an = Energy[i]
        
plt.plot(tp,Energy)
plt.title('Energy vs time')
plt.xlabel('time')
plt.ylabel('Energy')
```

```python
AM = xp*vyp - yp*vxp
plt.plot(tp,AM)
plt.title('Angular momentum vs time')
plt.xlabel('time')
plt.ylabel('Angular momentum')
```
