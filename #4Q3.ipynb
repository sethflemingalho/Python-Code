```python
import numpy as np
import astropy as ap
import matplotlib.pyplot as plt
from astropy.constants import h
from astropy import constants as const
import scipy.constants as sconst
```

```python
#h = const.h
#h = 6.62607004**-34
h = sconst.hbar
w = 2*np.pi*5*(10**15)
c = h*w
#m = const.m_e
#m = 9.109**-31 
m = sconst.m_e
```

```python
xo = -2*(10**(-10))
x1 = 2*(10**(-10))
x = np.linspace(xo,x1,10000)
```

```python
e0 = (0.5*c)
e1 = (1.5)*c
e2 = (2.5)*c

# psi0 = ((((m*w)/(np.pi*h))**(1/4))*np.e**(-(m*w*x**2))/(2*h))
# psi1 = (((m*w)/(np.pi*h))**(1/4)*(((2*m*w)/h)**1/2))*x*np.e**(-(m*w*x**2)/(2*h))
# psi2 = (1/(h*np.sqrt(2)))*(((m*w)/(np.pi*h))**(1/4))*np.e**(-(m*w*(x**2))/(2*h))*(2*m*w*(x**2)-(h**2))

psi0 = (((m*w)/(np.pi*h))**(1/4))*np.exp(-(m*w*(x**2))/(2*h))
psi1 = (((m*w)/(np.pi*h))**(1/4))*(((2*m*w)/h)**(1/2))*x*np.exp(-(m*w*(x**2))/(2*h))
psi2 = (1/(h*np.sqrt(2)))*(((m*w)/(np.pi*h))**(1/4))*np.exp(-(m*w*(x**2))/(2*h))*(2*m*w*(x**2)-(h**2))



W0 = e0 +(10**(-6))*e2*psi0
W1 = e1 + (10**(-6))*e2*psi1
W2 = e2 +(10**(-6))*e2*psi2


SHOpot = 0.5*m*(w**2)*(x**2)

ArrOfOnes = np.ones(10000)

v0 = e0*ArrOfOnes
v1 = e1*ArrOfOnes
v2 = e2*ArrOfOnes
```

```python
plt.figure(figsize=(8,8))
plt.plot(x,W0, label='psi0')
plt.plot(x,W1, label='psi1')
plt.plot(x,W2, label='psi2')
plt.plot(x,v0, label='e0')
plt.plot(x,v1, label='e1')
plt.plot(x,v2, label='e2')
plt.plot(x,SHOpot, label='SHO')

plt.xlim(-2*10**-10,2*10**-10)
plt.ylim(0,1*10**-17)
plt.xlabel('x(m)')
plt.ylabel('V(J)')
plt.legend(loc=3)
plt.title('Harmonic Oscillator')

```
