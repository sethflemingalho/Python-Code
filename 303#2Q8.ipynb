```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import scipy
from scipy import stats
```

```python
starcount = 10000
size = 800
coord = (np.random.random((starcount,3))-0.5)*size #-0.5 to center on sol
Gaussdist = np.random.normal(4.64, 0.6, starcount)
```

```python
fig = plt.figure()
fig = fig.add_subplot(111, projection = "3d")
fig.scatter(coord[:,0],coord[:,1],coord[:,2],s=Gaussdist)
plt.title("10000 stars within 400 Pc")
```

```python
fig1 = plt.figure()

logx = np.logspace(1.2, np.log10(400), 20, base=10)
logy = np.logspace(0, np.log10(1000), 20, base=10)
mask = np.sqrt(np.sum(coord**2,axis=1))

y = [coord[mask<x].shape[0] for x in logx]

fig = fig1.add_subplot(111)
fig.plot(logx,y)
fig.set_yscale('log')
fig.set_xscale('log')
fig.set_title("Density vs Distance to Sol")
plt.show()
```

```python
ig, fig = plt.subplots()
n,bins,patches=fig.hist(Gaussdist,histtype='stepfilled', bins=1000, alpha=0.8, density = True)
plt.title("Histogram of Absolute Magnitude")
```

```python
mu, sigma = scipy.stats.norm.fit(Gaussdist)

best_line = scipy.stats.norm.pdf(bins, mu, sigma)

plt.plot(bins, best_line)
```
