```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ao = 1
y = 0
alpha1 = 0
alpha2 = np.pi/6
alpha3 = np.pi/4
alpha4 = np.pi/3
alpha5 = np.pi/2
alpha6 = np.pi
xr = np.linspace(-10,10,100)
zr = np.linspace(-10,10,100)
```

```python
x,z = np.meshgrid(xr,zr)
r = np.absolute(np.sqrt(x**2+y**2+z**2))

probability1 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha1))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)

probability2 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha2))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)

probability3 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha3))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)

probability4 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha4))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)

probability5 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha5))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)

probability6 = (((((1/(10*np.pi*ao**3))*((1-(r/2*ao)))**2))+(((z*np.cos(alpha6))/(20*np.pi*ao**4))*(1-(r/2*ao))))\
                +((z**2)/(160*np.pi*ao**5)))*np.exp(-r/ao)
```

```python
plt.imshow(probability1, extent=[-10,10,-10,10], cmap = 'afmhot')
plt.colorbar()
plt.clim(0,0.0014)
plt.title("Probability Density Plot Alpha1")
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```

```python
B.) As alpha the (relative phase) increases the probability density shifts to 0 then increases in the -z direction. It will then hit a limit and reverse 
directions and begin once again increasing towards symmetry in the z axis. 
```
